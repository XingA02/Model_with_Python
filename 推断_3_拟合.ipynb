```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 生成一些示例数据
np.random.seed(0)
X = np.linspace(0, 10, 50)
y = 2.5 * X + np.random.normal(size=X.size)

# 线性拟合函数
def linear_func(x, a, b):
    return a * x + b

# 多项式拟合函数（假设二次多项式）
def polynomial_func(x, a, b, c):
    return a * x**2 + b * x + c

# 对数拟合函数
def logarithmic_func(x, a, b):
    return a * np.log(x) + b

# 指数拟合函数
def exponential_func(x, a, b):
    return a * np.exp(b * x)

# 使用curve_fit进行拟合
popt_linear, _ = curve_fit(linear_func, X, y)
popt_poly, _ = curve_fit(polynomial_func, X, y)
popt_log, _ = curve_fit(logarithmic_func, X, y)
popt_exp, _ = curve_fit(exponential_func, X, y)

# 绘制原始数据
plt.figure(figsize=(15, 5))
plt.scatter(X, y, label='Data')

# 绘制拟合曲线
plt.plot(X, linear_func(X, *popt_linear), 'r-', label='Linear Fit')
plt.plot(X, polynomial_func(X, *popt_poly), 'g--', label='Polynomial Fit')
plt.plot(X, logarithmic_func(X, *popt_log), 'b-.', label='Logarithmic Fit')
plt.plot(X, exponential_func(X, *popt_exp), 'm:', label='Exponential Fit')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Curve Fitting Examples')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
